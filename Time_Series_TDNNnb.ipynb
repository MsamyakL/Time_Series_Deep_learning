```
import pandas as pd
import numpy as np

import keras 
from keras.models import Sequential
from keras.layers import Conv1D, LSTM, Dense,Reshape,MaxPool1D
```

```
time_series = np.array([0,1,3,2,5,1,4,6,9,5,4,5,2,1,5,7,2,1,8,4,4,7,5,3])
```

```
time_series[0:2]
```

```
def split_seq(seq,n):
  X,y = list(), list()
  for i in range (len(seq)):

    end_of_seq = n + i 
    if end_of_seq >len(seq)-1:
      break
    seq1 = seq[i:end_of_seq]
    seq2 = seq[end_of_seq]
# Add stuff to the list
    X.append(seq1)
    y.append(seq2)


  return np.array(X),np.array(y)
    
```

```
split_seq(time_series,2)
```

```
X,y = split_seq(time_series, n=3)
```
